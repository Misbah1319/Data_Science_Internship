```python
import pandas as pd

# Load the Excel file into a pandas DataFrame
df = pd.read_excel('/Dataset .xlsx')

# Display the first 5 rows of the DataFrame
display(df.head())
```

```python
#Identifying the numbr of rows and columns in given  dataset
df.head()
df.shape
```

```python
# to check and handle the missing values
df.isnull().sum()
df.dtypes
df.fillna(df.mean(numeric_only=True),inplace=True)
df.fillna("Not Available",inplace=True)
```

```python
#Distribution of Aggregate rating
df["Aggregate rating"].dtype
df["Aggregate rating"]=df["Aggregate rating"].astype(float)
df["Aggregate rating"].value_counts()
df["Aggregate rating"].hist(bins=10)
```
