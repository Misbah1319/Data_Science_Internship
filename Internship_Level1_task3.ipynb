```python
import pandas as pd

# Load the Excel file into a pandas DataFrame
df = pd.read_excel('/content/Dataset .xlsx')

# Display the first 5 rows of the DataFrame
display(df.head())
```

```python
#Visualize restaurant locations
df.plot.scatter(x="Longitude", y= "Latitude")
```

```python
#Distributions across the cities & countries
df["Country Code"].value_counts()
df["City"].value_counts().head(10)
display(df["Country Code"].value_counts())
display(df["City"].value_counts().head(10))
```

```python
#Co-relation between location and rating
df.plot.scatter(x="Latitude", y="Aggregate rating")
df.plot.scatter(x="Longitude", y="Aggregate rating")
```
