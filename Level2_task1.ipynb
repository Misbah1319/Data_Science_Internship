```python
import pandas as pd
df=pd.read_excel('/content/Dataset .xlsx')
df.head()
```

```python
#Count total number if restaurants (total rows)
total_restaurants=len(df)
print("Total number of restaurants:",total_restaurants)
```

```python
#Filters restaurants that have table booking ='Yes'
table_booking_yes=df[df['Has Table booking']=='Yes'].shape[0]
```

```python
#Calculate percentage of restaurants offering table booking
table_booking_percentage=(table_booking_yes/total_restaurants)*100
```

```python
#Filters restaurants that have online delivery ='Yes'
online_delivery_yes=df[df['Has Online delivery']=='Yes'].shape[0]
#Calculate percentage of restaurants offering online delivery
online_delivery_percentage=(online_delivery_yes/total_restaurants)*100
print("Percentage of restaurants offering table booking:",table_booking_percentage)
print("Percentage of restaurants offering online delivery:",online_delivery_percentage)
```

```python
#Compare average ratings with vs without table booking
#select retaurants that provide table booking
with_table_booking=df[df['Has Table booking']=='Yes']
```

```python
#calculate average rating for restaurants with table booking
avg_rating_with_booking=with_table_booking['Aggregate rating'].mean()
```

```python
#Select restaurants that do not provide table booking
without_table_booking=df[df['Has Table booking']=='No']
#Calculate average rating for restaurants without table booking
avg_rating_with_booking=without_table_booking['Aggregate rating'].mean()
print("Average rating for restaurants with table booking:",avg_rating_with_booking)
print("Average rating for restaurants without table booking:",avg_rating_with_booking)
```

```python
#Display unique price range values
df['Price range'].unique()
#Group restaurants by price range & online delivery availability
price_range_delivery=df.groupby(['Price range','Has Online delivery']).size()
#Convert grouped data into a clean table
price_range_delivery=price_range_delivery.reset_index(name='Restaurant Count')
print(price_range_delivery)
```
