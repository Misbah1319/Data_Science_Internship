```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_excel('/content/Dataset .xlsx')
df.head()
```

```python
#Count how many restaurants fall under each price range
price_range_count=df['Price range'].value_counts()

#Display the count of each price range
price_range_count
```

```python
#Group restaurants based on price range
#and calculate the average rating for each group
avg_rating_price=df.groupby('Price range')['Aggregate rating'].mean()
#Display average rating for each price range
avg_rating_price
```

```python
#Group data based on rating color
#and calculate avg rating for each color
avg_rating_color=df.groupby('Rating color')['Aggregate rating'].mean()
#Display average rating for each color
avg_rating_color
```

```python
#Find the rating color with the highest average rating
highest_avg_rating_color=avg_rating_color.idxmax()
highest_avg_rating_color
```

```python
#For clear visualization
plt.xlabel("Price range")
plt.ylabel("Average rating")
plt.title("Average rating for each price range")
plt.bar(price_range_count.index,price_range_count.values)
plt.show()
```
