```python
import pandas as pd
df=pd.read_excel("Dataset .xlsx")
df.head()
```

```python
#New Feature : Restaurant_Name_Length
#Create a new feature that stores the length of each restaurant name
df['Restaurant_Name_Length']=df['Restaurant Name'].str.len()
#Display restaurant name and its length
df[['Restaurant Name','Restaurant_Name_Length']].head()
```

```python
#New feature:Address_Length
#Create a new feature that stores the length of each address
df['Address_Length']=df['Address'].str.len()
#Display address and its length
df[['Address','Address_Length']].head()
```

```python
#Encode Yes and No values into numeric format
#Yes-1(Table booking available)
#No-0(Table booking not available )
df['Has_Table_Booking']=df['Has Table booking'].map({'Yes':1,'No':0})
df[['Has Table booking','Has_Table_Booking']].head()
```

```python
df['Has_online_Delivery']=df['Has Online delivery'].map({'Yes':1,'No':0})
df[['Has Online delivery', 'Has_online_Delivery']].head()
df.info()
```
